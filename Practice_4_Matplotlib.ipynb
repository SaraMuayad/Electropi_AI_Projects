6- In the bigger plot draw Z based on x and in the smaller plot draw y based on x. Search on google to see how you can zoom inside a curve as you can see below (you should set xlim and ylim, search them in google)

```python

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
x = np.linspace(0, 100, 100)
Z = x**2

ax1.plot(x, Z)
ax1.set_xlabel('x')
ax1.set_ylabel('Z')

ax2.plot(x, Z)
ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])

plt.show()
```

8- Now plot y based on x and z based on x and two different subplots with the color and style shown below. don't forget to call fig to see the curves.

```python

fig, axes = plt.subplots(nrows=1, ncols=2)
x = np.linspace(0, 100, 100)
y1 = 2 * x
y2 = x**2
axes[0].plot(x, y1, color='blue', linestyle='--',linewidth=3)

axes[1].plot(x, y2, color='red', linestyle='-', linewidth=3)
plt.tight_layout()
plt.show()
```

9- Draw again the figure with the following size and attributes

```python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
x = np.linspace(0, 100, 100)
y1 = 2 * x
y2 = x**2
axes[0].plot(x, y1, color='blue', linestyle='-', linewidth=3)

axes[1].plot(x, y2, color='red', linestyle='--', linewidth=3)

plt.tight_layout()

plt.show()
```
